```python
movies['overview'].fillna('Null',inplace=True)
```

```python
#Now we will find similar movies just based on overview
#create a tfidf vectorizer for these movies


from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000,max_df=0.7)
x = vectorizer.fit_transform(movies['overview'])
tfidf_df = pd.DataFrame(x.toarray(), columns=vectorizer.get_feature_names())
print (x.toarray())
print (x.shape)
```

```python
#concatenate the dataframes
concat_df=pd.concat([movies[['original_title']],tfidf_df],axis=1)
```
